# ルンゲ・クッタ法の絶対安定領域

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set('notebook', 'whitegrid', 'dark', font_scale=1.5, rc={"lines.linewidth": 2, 'grid.linestyle': '--'})
```

```python
def phi1(z):
    return 1 + z

def phi2(z):
    return 1 + z + 0.5 * z**2

def phi3(z):
    return 1 + z + 0.5*z**2 + z**3/6

def phi4(z):
    return 1 + z + 0.5 *z**2 + z**3/6 + z**4/24
```

```python
x = np.linspace(-4,2,256)
y = np.linspace(-3,3,256)
X, Y = np.meshgrid(x, y)
Z1 = abs(phi1(X+1.j*Y))
Z2 = abs(phi2(X+1.j*Y))
Z3 = abs(phi3(X+1.j*Y))
Z4 = abs(phi4(X+1.j*Y))
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
ax.set_xlabel("$\mathrm{Re}(z)$")
ax.set_ylabel("$\mathrm{Im}(z)$")
ax.set_xlim(-4.5,2.5)
ax.set_ylim(-3.5,3.5)
c1 = ax.contour(X, Y, Z1,levels=[1.0], colors='k', linestyles='-')
c2 = ax.contour(X, Y, Z2,levels=[1.0], colors='k', linestyles='--')
c3 = ax.contour(X, Y, Z3,levels=[1.0], colors='k', linestyles='-.')
c4 = ax.contour(X, Y, Z4,levels=[1.0], colors='k', linestyles=':')
labels = ["$s=1$","$s=2$","$s=3$","$s=4$"]
c1.collections[0].set_label(labels[0])
c2.collections[0].set_label(labels[1])
c3.collections[0].set_label(labels[2])
c4.collections[0].set_label(labels[3])
plt.legend(loc='center right')
# plt.savefig("stability_RK1.pdf", bbox_inches='tight')
```

```python
def phi45(z):
    return 1 + z + 0.5 *z**2 + z**3/6 + z**4/24 + z**5/104

def phi54(z):
    return 1 + z + 0.5 *z**2 + z**3/6 + z**4/24 + z**5/120 + z**6/600
```

```python
x = np.linspace(-5,4,256)
y = np.linspace(-4.5,4.5,256)
X, Y = np.meshgrid(x, y)
Z4 = abs(phi4(X+1.j*Y))
Z45 = abs(phi45(X+1.j*Y))
Z54 = abs(phi54(X+1.j*Y))
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
ax.set_xlabel("$\mathrm{Re}(z)$")
ax.set_ylabel("$\mathrm{Im}(z)$")
ax.set_xlim(-5,4)
ax.set_ylim(-4.5,4.5)
c1 = ax.contour(X, Y, Z4,levels=[1.0], colors='k', linestyles='-')
c2 = ax.contour(X, Y, Z45,levels=[1.0], colors='k', linestyles='--')
c3 = ax.contour(X, Y, Z54,levels=[1.0], colors='k', linestyles=':')
labels = ["$s=4$", "RKF45", "DOPRI5"]
c1.collections[0].set_label(labels[0])
c2.collections[0].set_label(labels[1])
c3.collections[0].set_label(labels[2])
plt.legend(loc='center right')
# plt.savefig("stability_RK2.pdf", bbox_inches='tight')
```
