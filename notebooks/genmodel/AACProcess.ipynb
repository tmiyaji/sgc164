# $A+A \to C$ Process

```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set('poster', 'whitegrid', 'dark', rc={"lines.linewidth": 2, 'grid.linestyle': '-'})
import warnings
warnings.filterwarnings('ignore')
```

## 一度だけシミュレーション

```python
#rd.seed(20200801)
beta = 1e-8
N = 100000
a = np.zeros(N+1, dtype=int)
a[0] = 10000
for n in range(N):
    a[n+1] = a[n] - (1 if rd.random() < beta*a[n]*a[n] else 0)
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.plot(a, '-')
```

## 10回計算する

```python
def quad(beta):
    N = 100000
    a = np.zeros(N+1, dtype=int)
    a[0] = 10000
    for n in range(N):
        a[n+1] = a[n] - (1 if rd.random() < beta*a[n]*a[n] else 0)
    return a
```

```python
a = np.array([quad(beta) for k in range(10)])
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
for k in range(10):
    ax.plot(a[k,:])
```

## 平均をプロット

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.plot(np.mean(a, axis=0))
```

片対数プロット

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.plot(np.log(np.mean(a, axis=0)))
```

## 分散

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.plot(np.var(a, axis=0))
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.plot(np.log(np.var(a, axis=0)))
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.plot(np.std(a, axis=0))
```
