# モンテカルロ積分

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set('poster', 'whitegrid', 'dark', font_scale=1, rc={"lines.linewidth": 2, 'grid.linestyle': '-'})
```

```python
import random as rd
```

```python
def f(x):
    return 4/(1 + x*x)
```

numpy.random使用

```python
%%time
N = 100000000
Y = f(np.random.random(N))
print(np.mean(Y))
```

random使用

```python
%%time
N = 100000000
I = 0
for n in range(N):
    I += f(rd.random())
print(I/N)
```

誤差の減衰を観察

```python
nn = [10**k for k in range(8)]
for n in nn:
    print(n, abs(np.pi-np.mean(Y[:n])))
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(nn, [abs(np.pi-np.mean(Y[:n])) for n in nn], '-o')
ax.plot(nn, 1/np.sqrt(nn))
```
