# ２次元および３次元ブラウン運動

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
%matplotlib inline
sns.set('poster', 'whitegrid', 'dark', font_scale=1, rc={"lines.linewidth": 2, 'grid.linestyle': '-'})
```

## ２次元ブラウン運動

```python
h = 2**(-10)
N = 2**12
dW1 = np.insert(np.random.normal(loc=0.0, scale=h, size=N), 0, 0.0)
dW2 = np.insert(np.random.normal(loc=0.0, scale=h, size=N), 0, 0.0)
W1 = np.cumsum(dW1)
W2 = np.cumsum(dW2)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
ax.plot(W1, W2, '-')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
# plt.savefig("BrownianMotion2D.pdf", bbox_inches='tight')
```

## 3次元ブラウン運動

```python
h = 2**(-10)
N = 2**12
dW31 = np.insert(np.random.normal(loc=0.0, scale=h, size=N), 0, 0.0)
dW32 = np.insert(np.random.normal(loc=0.0, scale=h, size=N), 0, 0.0)
dW33 = np.insert(np.random.normal(loc=0.0, scale=h, size=N), 0, 0.0)
W31 = np.cumsum(dW31)
W32 = np.cumsum(dW32)
W33 = np.cumsum(dW33)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(W31, W32, W33,'-')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
# plt.savefig("BrownianMotion3D.pdf", bbox_inches='tight')
```
